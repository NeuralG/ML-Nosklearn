```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(42)
```

```python
import seaborn as sns
import random

df = sns.load_dataset("iris")
X = df.drop(columns="species")
y = df["species"] == "virginica"

indices = random.sample(range(len(X)), k=50)

X_train = X.drop(index=indices)
X_test = X.iloc[indices]

y_train = y.drop(index=indices)
y_test = y.iloc[indices]

X_train.insert(0, "bias", 1)
X_test.insert(0, "bias", 1)
```

```python
def computeGredient(X: pd.DataFrame, y: pd.Series, w: pd.Series):
    n = len(y)

    predictions = np.exp(X @ w) / (1 + np.exp(X @ w))
    error = predictions - y

    gradient = (1 / n) * X.T @ error

    return gradient


def gradientDescent(X, y, lr, iterCount):
    n = X.shape[1]
    coefficents = np.ones(n)

    for _ in range(iterCount):
        gradients = computeGredient(X, y, coefficents)
        coefficents -= lr * gradients

    return coefficents


finalW = gradientDescent(X_train, y_train, 0.1, 10000)

y_proba = predictions = np.exp(X_test @ finalW) / (1 + np.exp(X_test @ finalW))
y_preds = y_proba > 0.5

accuracy = np.mean(y_preds == y_test)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
