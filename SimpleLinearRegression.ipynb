# Simple Linear Regression with gredient descent


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
x = np.random.random(100)
y = x * 2 + np.random.random(100) / 5  # noise
```

```python
def lossFunctionDerivativeA(a, b, x, y):
    # f(x) = ax + b is regression line
    preds = a * x + b
    return -2 * np.mean(x * (y - preds))


def lossFunctionDerivativeB(a, b, x, y):
    preds = a * x + b
    return -2 * np.mean(y - preds)


lossFunctionDerivativeA(1, 0, x, y)
```

```python
def gredientDescent(lr, iterCount):
    a = 1.0
    b = 0.0

    for _ in range(iterCount):
        gradA = lossFunctionDerivativeA(a, b, x, y)
        gradB = lossFunctionDerivativeB(a, b, x, y)

        a -= lr * gradA
        b -= lr * gradB

    return a, b


a, b = gredientDescent(0.01, 10000)
a, b
```

```python
plt.scatter(x, y)
xs = np.linspace(x.min(), x.max(), 1000)
plt.plot(xs, xs * a + b, c="black", linewidth=3)
```
